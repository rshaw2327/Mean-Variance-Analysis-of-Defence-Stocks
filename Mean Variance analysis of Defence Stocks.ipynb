```python
import pandas as pd
import numpy as np
pd.options.display.float_format = '{:4f}'.format
import matplotlib.pyplot as plt
plt.style.use("seaborn")
import yfinance as yf
```

```python
start="2015-10-01"
end="2025-11-01"
```

```python
symbol=["LMT","RTX","NOC","GD","BA","LHX","HII","AIR"]
```

```python
df=yf.download(symbol,start,end,multi_level_index=True)
df
```

```python
df.info()
```

```python
df.to_csv("Defence_compaines_stocks")
```

```python
pd.read_csv("Defence_compaines_stocks",header=[0,1],index_col=0,parse_dates=[0])
```

```python
df.Close
```

```python
close=df.Close
```

```python
close.plot(figsize=(15,8),fontsize=13)
plt.legend(fontsize=13)
plt.show()
```

```python
print(close.head())
print(close.columns)
```

```python
log_returns = np.log(close / close.shift(1))
log_returns
```

```python
log_returns.columns = [f"{c}_log_return" for c in log_returns.columns]
```

```python
log_returns
```

```python
prices = close.copy()
log_returns = np.log(prices / prices.shift(1))
```

```python
summary=log_returns.agg(["mean","std"]).T # Transpose so we can get a column instead of a row
summary
```

```python
summary.plot(kind = "scatter",x="std",y="mean",figsize=(15,12),s=50,fontsize=15)
for i in summary.index:
    plt.annotate(i,xy=(summary.loc[i,"std"]+0.00005,summary.loc[i,"mean"]+0.00005),size=15)
plt.xlabel("Risk (std)",fontsize=15)
plt.ylabel("Mean Return",fontsize=15)
plt.title("Mean-Variance Analysis",fontsize=20)
plt.show()
```

```python
import matplotlib.pyplot as plt

ax = summary.plot(
    kind="scatter",
    x="std",
    y="mean",
    figsize=(15, 12),
    s=50
)

for asset in summary.index:
    ax.annotate(
        asset,
        xy=(summary.loc[asset, "std"], summary.loc[asset, "mean"]),
        xytext=(5, 5),
        textcoords="offset points",
        fontsize=15
    )

ax.set_xlabel("Risk (Standard Deviation)", fontsize=15)
ax.set_ylabel("Mean Return", fontsize=15)
ax.set_title("Mean–Variance Analysis", fontsize=20)

plt.show()
```
